```python
import cv2
import time 
import os
```

```python
facecap=cv2.CascadeClassifier("C:/Users/user/AppData/Local/Programs/Python/Python312/Lib/site-packages/cv2/data/haarcascade_frontalface_default.xml")
eyecap=cv2.CascadeClassifier("C:/Users/user/AppData/Local/Programs/Python/Python312/Lib/site-packages/cv2/data/haarcascade_eye.xml")

save_dir="saved_faces"
if not os.path.exists("saved_faces"):
    os.makedirs("saved_faces",exist_ok=True)

videocap= cv2.VideoCapture(0)
face_id=0
print("press 's' to save face | press 'q' to quit")

while True:
    ret,videodata=videocap.read()
    col=cv2.cvtColor(videodata,cv2.COLOR_BGR2GRAY)

    faces=facecap.detectMultiScale(
        col,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30,30),
        flags=cv2.CASCADE_SCALE_IMAGE
    )
    roi_color=None

    for(x,y,w,h) in faces:
        cv2.rectangle(videodata,(x,y),(x+w,y+h),(0,255,0),2)
    
        roi_gray= col[y:y+h, x:x+w]
        roi_color=videodata[y:y+h, x:x+w]

        
        eyes=eyecap.detectMultiScale(
            roi_gray,
            scaleFactor=1.1,
            minNeighbors=4
        )
        for(ex,ey,ew,eh) in eyes:
            cv2.rectangle(
                roi_color,
                (ex,ey),
                (ex+ew,ey+eh),(255,0,0),2
            )


    cv2.putText(videodata,"press s to save | q to quit",(20,40),
                cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,255),2
                )  
    
    cv2.imshow("FACE DETECTION",videodata)
    key= cv2.waitKey(10) & 0xFF

    if key == ord('s') and roi_color is not None:
        image_path = f"{save_dir}/face_{face_id}.jpg"
        cv2.imwrite(image_path,roi_color)
        print("saved:" ,image_path)
        face_id+=1
        time.sleep(0.3)

    if key == ord('q'):
        break

videocap.release()  
cv2.destroyAllWindows() 
```
